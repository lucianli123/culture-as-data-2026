## 1. Names and strings

Predict what prints, then run.

```python
line = "you said forever and i believed you"
title = line.upper()
print(title)
print(len(line))          # characters, not words
print(line.split()[:3])   # split makes a list of words; [:3] takes the first three
```

## 2. Numbers, and a small surprise

Predict all three lines.

```python
plays = 1_000_000
share = 34.1
print(plays * 2)
print(share / 100)
print(7 // 2, 7 % 2)      # whole-number division, and the remainder
```

## 3. A list and a loop

Before running: how many lines will print, and what will the last one be?

```python
verbs = ["snuggles", "giggles", "gallops", "shoots"]
for v in verbs:
    print(v, len(v))
print("the last one was:", verbs[-1])   # -1 means "count from the end"
```

## 4. A dict: labeled slots

APIs hand you these. Predict what each line shows.

```python
film = {"title": "The Red Shoes", "year": 1948, "rating": 4.5}
print(film["title"])
print(film.get("director", "not recorded"))   # .get gives a default instead of an error
film["seen"] = True
print(film)
```

## 5. A function: a recipe with a name

Nothing happens until it's *called*. Predict the two outputs.

```python
def brightness(r, g, b):
    return 0.299 * r + 0.587 * g + 0.114 * b   # the same formula as the Week 1 pixel lab

print(brightness(255, 255, 255))   # white
print(brightness(20, 20, 40))      # the "midnight" swatch
```

## 6. A tiny DataFrame

The object most course code passes around. Predict the shape, then the counts.

```python
import pandas as pd

df = pd.DataFrame([
    {"artist": "A", "word": "forever", "count": 12},
    {"artist": "A", "word": "coast",   "count": 5},
    {"artist": "B", "word": "forever", "count": 2},
])
print(df.shape)                          # (rows, columns)
print(df.groupby("artist")["count"].sum())
```
