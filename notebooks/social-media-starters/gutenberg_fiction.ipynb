```python
import io, re
import pandas as pd

EXCERPT = ("It was on a dreary night of November that I beheld the accomplishment "
           "of my toils. With an anxiety that almost amounted to agony... " * 40)

def load_book(gutenberg_id=84, title="Frankenstein"):
    """Fetch a plain-text book by Gutenberg ID; on any failure, a built-in excerpt."""
    try:
        import requests
        url = f"https://www.gutenberg.org/files/{gutenberg_id}/{gutenberg_id}-0.txt"
        text = requests.get(url, timeout=30).text
        body = re.split(r"\*\*\* ?START OF (?:THE|THIS) PROJECT GUTENBERG.*?\*\*\*", text)[-1]
        body = re.split(r"\*\*\* ?END OF (?:THE|THIS) PROJECT GUTENBERG", body)[0]
        return title, body
    except Exception as e:
        print("no network, using the excerpt:", e)
        return title, EXCERPT

title, text = load_book(84, "Frankenstein")
print(title, "-", len(text.split()), "words")
```

```python
# Find more: the Gutendex search API (https://gutendex.com/books?search=...) returns
# IDs by title, author, or subject ("science fiction", "horror tales"). A genre corpus
# is a list of IDs and a loop over load_book.
GOTHIC = {84: "Frankenstein", 345: "Dracula", 43: "Jekyll and Hyde"}
corpus = {t: load_book(gid, t)[1] for gid, t in GOTHIC.items()}
print(len(corpus), "books loaded")
```
