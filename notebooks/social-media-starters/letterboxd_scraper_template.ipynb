# Scraper template: Letterboxd (and the Goodreads shortcut)

Letterboxd has no official API; **light scraping with delays is feasible at small
scale**. Goodreads needs no scraper at all — the **UCSD Book Graph** is the standard
academic dump; download it instead.

This is a *template*: the parsing runs on a built-in page so it works offline; point it
at real pages only after the checklist below.

**The polite-scraping checklist** (from `kits/licensing-one-pager.md`):

1. Read `robots.txt` and the terms of service first.
2. One request every few seconds; identify yourself in the user agent.
3. Take only what the question needs; stop when you have it.
4. Never republish scraped text; publish counts and findings.

```python
import time
import pandas as pd
from bs4 import BeautifulSoup

# The template parses this built-in page; a real film page has the same shape.
PAGE = """
<html><body>
  <ul class="film-list">
    <li class="film"><span class="name">The Red Shoes</span><span class="rating">4.5</span><span class="year">1948</span></li>
    <li class="film"><span class="name">Suspiria</span><span class="rating">4.0</span><span class="year">1977</span></li>
    <li class="film"><span class="name">Neon Demon</span><span class="rating">3.5</span><span class="year">2016</span></li>
  </ul>
</body></html>
"""

def parse_page(html):
    soup = BeautifulSoup(html, "html.parser")
    return pd.DataFrame([{
        "name": li.select_one(".name").get_text(),
        "rating": float(li.select_one(".rating").get_text()),
        "year": int(li.select_one(".year").get_text()),
    } for li in soup.select("li.film")])

def fetch(url):
    """Swap in for real pages AFTER the checklist. Keeps the delay for you."""
    import requests
    time.sleep(3)                                   # the polite pause, non-negotiable
    return requests.get(url, timeout=30,
                        headers={"User-Agent": "culture-as-data student project"}).text

films = parse_page(PAGE)
films
```

**Adapting the template:** open the real page in your browser, View Source, find the
repeating element that holds each row, and change the three `select_one` selectors. That
is the whole craft; the AI assistant is good at writing selectors if you paste it a
snippet of the real HTML.
